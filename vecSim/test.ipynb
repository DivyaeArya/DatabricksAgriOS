```python
from databricks.vector_search.client import VectorSearchClient

# Initialize the Vector Search client
vsc = VectorSearchClient(
     workspace_url="https://dbc-a6b1b203-95b7.cloud.databricks.com",
     personal_access_token="dapi9ae9552ee49526db039a9f8dff3eb100"
)

# Configuration
endpoint_name = "kisanqna"
index_name = "workspace.default.kisanindex"

# Test queries for farming
test_questions = [
    "How to control pests in crops?",
    "What is the best fertilizer for wheat?",
    "How to increase crop yield?"
]

print("=" * 80)
print("VECTOR SIMILARITY SEARCH TEST - KisanIndex")
print("=" * 80)

# Get the index
index = vsc.get_index(
    endpoint_name=endpoint_name,
    index_name=index_name
)

# Check index status first
status = index.describe()
index_status = status.get('status', {}).get('state', status.get('index_status', 'ONLINE'))
print(f"\nIndex Status: {index_status}")
print(f"Indexed rows: {status.get('num_indexed_rows', 0)}")
print()

# Test each query
for i, query in enumerate(test_questions, 1):
    print(f"\n{'='*80}")
    print(f"Query #{i}: {query}")
    print('='*80)
    
    # Perform similarity search
    results = index.similarity_search(
        query_text=query,
        columns=["id", "questions", "answers"],
        num_results=3
    )
    
    # Display results - results are returned as lists
    if results and 'result' in results and 'data_array' in results['result']:
        for j, result in enumerate(results['result']['data_array'], 1):
            # Result format: [id, questions, answers, similarity_score]
            if isinstance(result, list) and len(result) >= 4:
                result_id = result[0]  # ID (numeric)
                question = result[1]    # Question text
                answer = result[2]      # Answer text
                score = result[3]       # Similarity score
                print(f"\n--- Result {j} (Similarity Score: {score:.4f}) ---")
                print(f"ID: {result_id}")
                print(f"Question: {question}")
                # Handle answer - convert to string and truncate
                answer_str = str(answer) if answer is not None else 'N/A'
                print(f"Answer: {answer_str[:300]}..." if len(answer_str) > 300 else f"Answer: {answer_str}")
            else:
                print(f"\n--- Result {j} ---")
                print(f"Data: {result}")
    else:
        print("⚠️ No results found")

print(f"\n{'='*80}")
print("✓ Test completed!")
print('='*80)
```
